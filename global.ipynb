## Vi

```python
for column in df:
    plt.figure(figsize=(9,5))
    plt.hist(df[column] , edgecolor= "black")

    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel("Frequency")

    plt.show()
```

```python
for col in df_num:
    plt.figure(figsize=(10,6))
    df[col].plot(kind='bar')

    plt.title(f'Bar Chart of {col}')
    plt.xlabel("Country")
    plt.ylabel(col)
    plt.tight_layout()

    plt.show()
```

### Energy Consumpion Trend Over the Years:

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df , x='Year' , y='Total Energy Consumption (TWh)', hue='Country' , alpha=0.6)

plt.title('Total Energy Consumion Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Energy Consumption (TWh)')
plt.legend([],[], frameon=False)
plt.show()
```

### Top 10 countries With Highest Energy Consumption:

```python
top_countries = df.groupby("Country")["Total Energy Consumption (TWh)"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar' , color='orange')

plt.title('Top 10 Countries by Energy Consumption')

plt.ylabel('Total Energy Consumption (TWh)')
plt.xticks(rotation=45)
plt.show()
```

### Renewable vs No-Renewable Energy Usage:

```python
plt.Figure(figsize=(12,6))
sns.histplot(df['Renewable Energy Share (%)'] ,bins=20 , kde=True , color='green')

plt.title('Distribution of Renewable Energy Usage (%)')
plt.xlabel('Renewable Energy Share (%)')
plt.show()
```

### Correlation Matrix:

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr() , annot=True , cmap='coolwarm' , fmt='.2f' , linewidths=0.5)

plt.title('Correlation Matrix of Features')
plt.show()
```
